```python
# condicionales anidados
# Condicionales con if elif y else
edad = int(input('Dime tu edad'))
if(edad<18):
    print("Menor de edad")
elif(edad >= 18 and  edad<65):
    print("En edad de trabajar")
    if(edad>= 18 and edad<45):
        print("Edad media 1")
    elif (edad>= 45 and edad <65):
        print("Edad media 2")
else:
    print('Está jubiliado')
    print("Por fin")
```

```python
#While : bucle condicional
#break (salir del bucle)
# Pregunta al usuario por un número hasta que introduce 0.
while True:
    num= int(input('Introduce un número:'))
    if num ==0:
        print('Vamos a salir porque ya se ha metido el 0')
        break
    
```
