```python
numeros = [3, 7, 1, 9, 4]

def doble(n):
    return n * 2

resultado = list(map(doble, numeros))
print(resultado)  # [6, 14, 2, 18, 8]
```

```python
resultado = list(map(lambda n: n * 2, numeros))
print(resultado)  # [6, 14, 2, 18, 8]
```

```python
resultado = list(map(lambda n: n + 1, numeros))
print(resultado)  # [6, 14, 2, 18, 8]
```

```python
numeros = [3, 7, 1, 9, 4]

def doble(n):
    return n * 2

resultado = list(map(doble, numeros))
print(resultado)  # [6, 14, 2, 18, 8]

def mayor_que_10(n):
    return n > 10

filtrados = list(filter(mayor_que_10, resultado))
print(filtrados)  # [14, 18]
```

```python
filtrados = list(filter(lambda n: n > 10, map(lambda n: n * 2, numeros)))
print(filtrados)  # [14, 18]
```

```python
from functools import reduce
'''
1er paso:  a=3,  b=7  →  3 + 7  = 10
2do paso:  a=10, b=1  →  10 + 1 = 11
3er paso:  a=11, b=9  →  11 + 9 = 20
4to paso:  a=20, b=4  →  20 + 4 = 24
a siempre es el resultado acumulado del paso anterior
b siempre es el siguiente elemento de la lista
Al final solo queda un único valor
'''
numeros = [3, 7, 1, 9, 4]

resultado = reduce(lambda a, b: a + b, numeros)
print(resultado)  # 24
```

```python
resultado = reduce(lambda a, b: a * b, numeros)
print(resultado)  # 756
```
